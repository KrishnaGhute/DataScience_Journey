```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# for feature selection
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# to visualize all the columns in data frarme
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv("house_train.csv")
```

```python
dataset.head()
```

```python
# capture teh dependent feature
y_train=dataset[['SalePrice']]
```

```python
# drop dependent feature from dataset
x_train=dataset.drop(['Id','SalePrice'],axis=1)
```

```python
## apply feature selection
# first specify the Lasso Regerssionn model and 
# select a suitable alpha (equivalent of penalty)
# the bigger the alpha the less features that will be selected.

# then use the SelectFromModel object from sklearn, which
# will select the feture which coefficients are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # ramember to set the sead value
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
# let's print the number of total and select features

# this is how we can make a list of the selected features
selected_feat = x_train.columns[feature_sel_model.get_support()]

# printing some stats
print('total features: {}'.format(x_train.shape[1]))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrunk to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)
))

```
