# Ordinal numbering encoding or label encoding
## Ordinal categorial varialbes
Ordinal data is a catogorical, dtatisticla data type where the variables have natural, ordered categories and the distance between the categories is not known.

```python
import pandas as pd
import datetime
```

```python
# create a variable with dates, and form that extract the weekday
# creating a list of dates with 20 days different form today
# then transform it into a dataframe
df_base=datetime.datetime.today()
df_date_list=[df_base-datetime.timedelta(days=x) for x in range(0,20)]
df=pd.DataFrame(df_date_list)
df.columns=['day']
df
```

```python
# extract the week day name
df['day_of_week']=df['day'].dt.day_name()
df.head()
```

```python
# engineer catogericla varaibles by ordinal number replacement
weekday_map={'Monday':1,
             'Tuesday':2,
             'Wednesday':3,
             'Thursday':4,
             'Friday':5,
             'Saturday':6,
             'Sunday':7
                 }
df['day_ordinal']=df.day_of_week.map(weekday_map)
df.head(20)
```
