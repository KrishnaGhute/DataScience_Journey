# minmax sclaing
- minmax scaling scales the vales betweeen 0 to 1.
- x_scaled = (x - x.min ) / ( x.max - x.min)

### usages
- specially used in the CNN in deep learning techniques.


- but some peoples use it in machine learning also

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic_train.csv',usecols=['Survived','Pclass','Age','Fare'])
```

```python
min_max=MinMaxScaler()
# this time we will directly convert values into the dataframe
df_minmax=pd.DataFrame(min_max.fit_transform(df),columns=df.columns)
df_minmax.head()
# this time head() will work because dataframe
```

## Pclass

```python
# berore transformation 
plt.hist(df['Pclass'],bins=20)
```

```python
# after transformation 
plt.hist(df_minmax['Pclass'],bins=20)
```

# Age

```python
# berore transformation 
plt.hist(df['Age'],bins=20)
```

```python
# Age after transformation 
plt.hist(df_minmax['Age'],bins=20)
```

# Fare

```python
# beror transformation 
plt.hist(df['Fare'],bins=20)
```

```python
# after transformation 
plt.hist(df_minmax['Fare'],bins=20)
```

# results
- we can see almost every result get converted into 0 to 1 range
