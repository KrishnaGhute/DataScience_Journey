# One Hot Encoding - variavles with many catogeries

```python
import pandas as pd
import numpy as np

data=pd.read_csv('mercedesbenz.csv')
data.head()
```

```python
data.shape
```

```python
cat_variable=[feature for feature in data.columns if data[feature].dtype=='object']
cat_variable
```

```python
cat_data=data[cat_variable]
cat_data.head()
```

```python
# let's check how many unique labels are each variable have
for col in cat_data.columns:
    print(col, ' : ',len(cat_data[col].unique()),' labels')
```

```python
# lets examine how many columns we will obtain after one hot encoding these variables
pd.get_dummies(cat_data,drop_first=True).shape
```

we can see we ended form just 8 categorical variables to the 187 new varialbes

## what we can do

- Ia a compitation authors limit one hot encoding to the 10 most frequent labesls of the variable. this means that they would make one binary variabel for each of the 10 most frequent labels only. This is equivalent to grouping all the other lables under a new catogery, that in this case will be droped. Thus, 10 new dummy variables indicates if one of the 10 most ferquent labels is present (1) or not (0) for a particular observation.

## how we can obtain in python

```python
# lets find the top 10 most frequent catogeries for the variables x2
cat_data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# let's make a list with the most frequent categories of the variable
top_10=[x for x in cat_data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# now we maeke the 10 binary variables
for label in top_10:
    cat_data[label]=np.where(data['X2']==label,1,0)

cat_data[['X2']+top_10].head(40)
```

```python
# get whole set of dummy variables, for all the categoriacal varables
def one_hot_top_x(df,variable,top_x_labels):
    # function to create the dummy variables for the most frequent labels
    # we can vary the number of most frequent labels that we encode

    for label in top_x_labels:
        df[variable+'_'+label]=np.where(data[variable]==label,1,0)

# read the data again
data=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])

# encode X2 into the 10 most frequent catogeries
one_hot_top_x(data,'X2',top_10)
data.head()
```

```python
# find the 10 most frequent dummy variables for X1
top_10=[x for x in data.X1.value_counts().sort_values(ascending=False).head(10).index]

# now create the 10 most frequent dummy variables for X1
one_hot_top_x(data,'X1',top_10)
data.head()
```
