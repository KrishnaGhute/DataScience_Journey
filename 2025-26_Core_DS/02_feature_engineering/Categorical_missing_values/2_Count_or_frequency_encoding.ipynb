# Handle catogericla features many
# Count or frequency encoding
## High cardinality
- variables wiht multitude catogeries, are also called as high cardinality.
- if variable have multiple labels then we use one hot encoding, using this we expand the feature space dramatically
- one of the approach is heavily used in kaggle compititaitons, it to repalce each lable of the cateforicla variable by the count, this is the amount of thimes each label appears in the dataset. or the frequency, this is the percentage of bservation within that catogry. the 2 are equivallent.

```python
import pandas as pd
import numpy as np

df=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2'])
df.head()
```

```python
df.shape
```

# one hot encoding

```python
pd.get_dummies(df).shape
# we can see form 2 columns we are getting the 71 columns
```

```python
len(df['X1'].unique())
```

```python
len(df['X2'].unique())
```

```python
# lets take a look on how many labels are there
for col in df.columns[0:]:
    print(col,' : ',len(df[col].unique())," labels")
```

```python
# lets obtain the count for each one of the labels in vauable X2
# lets capture this in a dictionary that we can use to re-map the lablesfor 
df["X2"].value_counts().to_dict()
```

```python
# now lets replace each label in X2 by its count
# first we amek a dictionary that maps each label to the count
df_frequency_map=df.X2.value_counts().to_dict()
```

```python
df.X2.head()
```

```python
# we replace X2 labels in the dataset df
df.X2=df.X2.map(df_frequency_map)
df.head(100)
```

There are some adventages and disadventages that will discuss now
# adv
1. it is very simple to implement
2. Does not increase the future dimensional space
# disadv
1. if some of the labels have the same count, then they will be repaced wiht the same count and they will loose some valuable information
2. adds somewhat arbitary numbers, and therefore weights to the different labels, that may not be related to their predictive powers
