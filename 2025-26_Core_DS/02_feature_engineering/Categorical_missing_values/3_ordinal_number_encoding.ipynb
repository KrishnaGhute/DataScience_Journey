# Target Guided Ordinal Encoding
1. Ordering the labels according to target.
2. Replace the labels by the joint probability being 1 or 0

```python
import pandas as pd
df=pd.read_csv('titanic_train.csv', usecols=['Cabin','Survived'])
```

```python
df.head(10)
```

```python
df.fillna({'Cabin': 'Missing'})
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
# finding unique values
df.Cabin.unique()
```

```python
# finding mean of survived from cabin wiht unique catogeries
df.groupby(['Cabin'])['Survived'].mean()
```

```python
# sorting the mean values.
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
# using enumerate we will assing the ranking to the ordinal labels
# we are starting labeling with 0 because the model will get more information
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
# mapping a new column with the ordinal_label2
df['cabin_ordianl_labels']=df['Cabin'].map(ordinal_labels2)
```

```python
df.head()
```

# Mean encoding

```python
mean_ordianl=df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordianl
```

```python
df['mean_ordianl_encoding']=df['Cabin'].map(mean_ordianl)
```

```python
df.head()
```

# Adventages:
1. capture information in lable therefore rendring more predective features.
2. it creates the monotonic relationship between target and the variable.

# Disadventages:
1. it leads to overfitting

# Probability Ratio Encoding

```python
df=pd.read_csv('titanic_train.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
# replacing
df.fillna({'Cabin': 'Missing'}, inplace=True)
```

```python
df.head()
```

```python
df["Cabin"].unique()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df['Cabin'].unique()
```

```python
# with respect to cabin what is survival rate.
prob_df=df.groupby(['Cabin'])['Survived'].mean()
prob_df
```

```python
# probability of survived
prob_df=pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
# probability of death
prob_df["Died"]=1-prob_df["Survived"]
prob_df
```

```python
# formul for probabily ratio is probability of survived by probability of died.
prob_df["probability_ratio"]=prob_df['Survived']/prob_df['Died']
prob_df.head()
```

```python
# making probability_ratio a dictionary so we can replace nan values with it.
prob_encoded=prob_df['probability_ratio'].to_dict()
prob_encoded
```

```python
df['Cabin_encoded']=df['Cabin'].map(prob_encoded)
df.head()
```

- we can skip the cabin column and use cabin_encoded column for model training
