# Statistics Practical Implementation
## Measure of central tendency 
1. Mean
2. Median
3. Mode

```python
import pandas as pd
import numpy as np

ages = [18, 21, 23, 25, 27, 29, 30, 32, 34, 35, 23,
        36, 38, 40, 42, 45, 48, 50, 55, 60, 65,120]

ages=np.sort(ages)
print(ages)
```

```python
print(np.mean(ages))
print(np.median(ages))
```

```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
print(statistics.mode(ages))
```

```python
import seaborn as sns
sns.boxplot(ages)
```

## 5 Number summary

```python
q1,q3=np.percentile(ages,[25,75])
print(q1,q3)
```

```python
# checking outliers
IQR=q3-q1
lf=q1-1.5*(IQR)
hf=q3+1.5*(IQR)
print(f"lower fence {lf}")
print(f"higher fence {hf}")
```

## Measure of Dispersion
1. variance
2. standard deviation

```python
# statics.variance() uses the sample formula having (n-1)
print(statistics.variance(ages))
```

```python
# calculating the sample variance by manually
def variance(data):
    n=len(ages)
    # mean of the data
    mean=sum(data)/n
    # variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-1)
    return variance
```

```python
print(variance(ages))
```

```python
# finding the population variance by the statistics
print(statistics.pvariance(ages))
```

```python
# np.variance() uses the population formula having (n)
print(np.var(ages,axis=0))
```

```python
# calculating the population variance by manually
def variance(data):
    n=len(ages)
    # mean of the data
    mean=sum(data)/n
    # variance
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/n
    return variance
```

```python
print(variance(ages))
```

```python
# population variance through the maths
import math
math.sqrt(statistics.pvariance(ages))
```

## check whether the distribution is normal or not

```python
# If you want to check whether frature is gaussian or normal distributed
# Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
# craeate a normal distribution data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
# we can see a perfectly alligned line
```

```python
# for lognormal distribution it is not properly alligned
mu, sigma = 3. , 1.  # mean and stsandard deviation
s=np.random.lognormal(mu,sigma,1000)
plot_data(s)
```

```python
# we are checking for the converted lognormal to noraml
plot_data(np.log(s))
```

```python
# lets us check for sepal_width form iris dataset
# we had seen it has a normal distribution
plot_data(df['sepal_width'])
```
