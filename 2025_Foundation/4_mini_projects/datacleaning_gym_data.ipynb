# method 2

```python
# if we have multiple greater values then we can use conditional statements
df=pd.read_csv("fitness.csv")
for x in df.index:
    if df.loc[x,"Duration"]>70:
        df.loc[x,"Duration"]=50
df
```

```python
# if we want to remove all rows having greater values then
df=pd.read_csv("fitness.csv")
for x in df.index:
    if df.loc[x,"Duration"]>70:
        df.drop(x, inplace=True)
df
# we can see 7th row get removed
```
