# Eval() Function
## python eval function - Evaluating experssion dynamically
- Eval function evaluates python experssion which are written as strings

```python
# Normal experssion
3*8
```

```python
# using the eval function
eval("3*8")
```

```python
# we can give input to the eval function
eval(input("Enter the experssion:"))
```

```python
# using function through eval
def square_num(num):
    return num**3
```

```python
# normal approach
square_num(6)
```

```python
# using eval function
eval("square_num(6)")
```

## How does Eval works
1. Parse python experssion
2. compile into a byte code
3. Evaluate the python expression
4. It will return the result

```python
# compile function
var=compile("5*5","<string>","eval")
```

```python
var
```

```python
eval(var)
```

```python
# globals : it must be a dictaniory
eval("x+50*4/2.5",{"x":100})
```

```python
# or
x=10
eval("x*5+8/57-5",{"x":x})
```

```python
# or
x=100
z=30     # ans=180
eval("x+y+z",{"x":x,"z":70,"y":50})
```

```python
# boolean expression
a=200
b=100
eval("a>b")
```
