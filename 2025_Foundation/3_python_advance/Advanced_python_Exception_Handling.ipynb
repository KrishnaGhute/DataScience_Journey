# try else
- if we take print in else block then it gives succesfull terminations.
- if there is any exception then the else block does not get execuated.

```python
try:
    a=int(input("enter 1st number:"))
    b=int(input("enter 2nd number:"))
    c=a/b
    d=a*b
    f=a+b
except NameError:    
    print('The user have not defined the variable.')
except TypeError:
    print('Try to add similar data types.')
except ZeroDivisionError:
    print('Please provide number greater than zero')
except Exception as ex: # ex is called as alliase
    print(ex)
else:
    print(c)
    print(d)
    print(d)
```

# try else finally
- finally block get execuated in each condition.
- the closing of open databse or string after handling exception it is closed by finally.

```python
try:
    a=int(input("enter 1st number:"))
    b=int(input("enter 2nd number:"))
    c=a/b
    d=a*b
    f=a+b
except NameError:    
    print('The user have not defined the variable.')
except TypeError:
    print('Try to add similar data types.')
except ZeroDivisionError:
    print('Please provide number greater than zero')
except Exception as ex: # ex is called as alliase
    print(ex)
else:
    print(c)
    print(d)
    print(d)
finally:
    print('execuation is succcesfull.')
```

# Custom Exception

```python
# making an inheritance class for exception handling
class error(Exception):
    pass    # using pass b'coz I want to pass my own custom statements
class dobError(error):
    pass
```

```python
year=int(input("Enter your birth year:"))
age=2025-year
try:
    if age<=30 and age>=20:
        print('You are eligible for the exam.')
    else:
        raise dobError
except dobError:
    print('your age is not valid for the exam.')
```
