# Generators
- basically used to create iterators

```python
# using return
def square(n):
    for i in range(n):
        return i**2
# return iterate from 0 to 3 and give first element which is 0.
```

```python
square(3)
```

```python
# using yield instead of return.
def square(n):
    for i in range(n):
        yield i**2

x=square(4)
for values in x:
    print('positon:{} and value:{}'.format(x,values))
```

- it will make square function iteraor.
- But, actually it is a generator.
## return
- return gives the iteration through first element only.
## yield
- yield itrate through all elements and save all the elements.

```python
a=square(3)
a
```

```python
# we can access the iteraors by for loop.
for i in a:
    print(i)
```

```python
# we can also use the next()
# all elements are iterated by for loop thats why we get erroe.
# let us handle it by try and except
try:
   print(next(a))
except:
    print("all elements are already itered")
```

# Closure
- basically we are able to acces the variable of main function through the subfunction defined in main function.

```python
def main_welcome():
    # we can provide prameter in function but let us create a variable
    msg="Hi"

    def sub_welcome_class():
        print('Excuse me.')
        print(msg)
        print('I am krishna.')
    return sub_welcome_class()
```

```python
main_welcome()
```

```python
# closures and initial decorators
# We can pass the function defined in main function then we can access it in the sub function.
def main_welcome(msg):

    def sub_welcome_class():
        print('Excuse me.')
        print(msg([1,2,3,4,5]))
        print('I am krishna.')
    return sub_welcome_class()
```

```python
main_welcome(len)
```

# Decorators
- inside in inner function we can give parameter as function.

```python
# Decorator function
def main_welcome(msg):

    def sub_welcome_class():
        print('Excuse me.')
        msg()
        print('I am krishna.')
    return sub_welcome_class()
```

```python
def request():
    print("Are you buzzy ?")
```

```python
main_welcome(request)
# first request function accessed by main_welcome
# then it goes to msg() function defined in the defined function.
```

```python
# Anoter way 
# call decorator on the top of the function.
@main_welcome
def request():
    print("Are you buzzy ?")
```

## copy
shallow copy

```python
lst1=[1,2,3,4]
print(lst1)
lst2=lst1.copy()
lst1[1]=100    # making changes after copy dosn't affects the lists.
print(lst2,lst1)
```

```python
lst1=[1,2,3,4]
lst1[1]=100    # making changes before copy affects the lists.
print(lst1)
lst2=lst1.copy()
print(lst2,lst1)
```
