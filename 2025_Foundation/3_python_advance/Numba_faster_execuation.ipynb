## Numba
- in-time compiler for python works best for the arrays, function and loops.

```python
#!pip install numba
```

```python
from numba import jit
import numpy as np
import time

x=np.arange(100).reshape(10,10)

@jit(nopython=True)
def go_fast(a):      # function is compiled and run in machine code
    trace=0.0
    for i in range(a.shape[0]):
        trace+=np.tanh(a[i,i])
        return a+trace

# do not report this...   compilation time is included in the execuation time!
start=time.time()
go_fast(x)
end=time.time()
print("Elapsed (with Compilation)=%s" % (end-start))

# now the function is compiled, re-time it execuating from cache
start=time.time()
go_fast(x)
end=time.time()
print("Elapsed (after Compilation)=%s" % (end-start))
```
