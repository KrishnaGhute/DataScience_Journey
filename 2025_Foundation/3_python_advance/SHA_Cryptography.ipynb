```python
import hashlib
```

## SHA256

```python
str1="krishna"
# Process of Hashing
# First step is to encode
# Then apply hashing Algorithums
hashedval=hashlib.sha256(str1.encode())
print(hashedval)
```

```python
# converting this value in hexaadecimal
hashedval.hexdigest()
```

## SHA384

```python
hashedval=hashlib.sha384(str1.encode())
hashedval.hexdigest()
```

```python
hashedval=hashlib.sha512(str1.encode())
hashedval.hexdigest()
```
