# Instantiation: 
- Creating two separate instances (objects) from the Dog class.

```python
class car:
    def __init__(self,w,d,e):
        self.window=w    #instance variable
        self.door=d       #instance variable
        self.enginetype=e  #instance variable
    # here I am going to create a new mentod where i acces enginetype from above class
    def self_drive(self):
        return "this is a {} car".format(self.enginetype)

# Instantiation: Creating two separate instances (objects) from the Dog class.
car1=car(5,4,'v12')
car2=car(3,2,'peterol')

# while calling the objects write "instance.instance_variable"
print(f"The car have {car1.window} windows, {car1.door} and it have {car1.enginetype} enginetype.")
```

```python
print(car2.enginetype)
```

```python
# I am going to call values from car1
car1.self_drive()
```

# Inheritance

```python
# All the class variables are public
# Car Blueprint
class student():
    def __init__(self,r,m,n):
        self.roll_no=r
        self.marks=m
        self.name=n
    def remark(self):
        print('The student have excellent performance=a+.')
```

```python
std1=student(23,95,'krishna')
```

```python
std1.name
```

```python
print(std1.marks)
```

```python
std1.remark()
```

```python
class talents(student):
    def __init__(self,r,m,n,t):
        super().__init__(r,m,n)
        self.talent=t
    def showcase(self):
        print("He is a Gaming master.")
```

```python
std2=talents(24,95,"adibal","Gaming")
```

```python
print(std2.roll_no)
```

```python
print(std2.talent)
```

```python
std2.showcase()
```

# OOPS Magic Methods In Classes

```python
# using __str__ to override the above functions.
# use return in __str__ then print function can give the return statement.
class car():
    # the __new__ is initialized before the __init__ 
    def __new__(self,w,d,e):
        print('objects are initialiazing .....')
    def __init__(self,w,d,e):
        self.windows=w
        self.door=d
        self.enginetype=e
    # now str over ride above functions
    def __str__(self):
        return "The object has been initialized."
    # Override the __sizeof__
    def __sizeof__(self):
        return "The size is given by this." 
    def drive(self):
        print("The person drives the car.")
```

```python
c=car(4,5,"v12")
# we can see __new__ first get initialized.
```

```python
c
```

```python
# print function can access the return function
print(c)
```

```python
# we can check the memory space taken by c
c.__sizeof__()
```
