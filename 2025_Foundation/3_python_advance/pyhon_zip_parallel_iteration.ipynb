# zip function
- it allow to iterate through the multiple iterators parallely
- we can use for loop and next to iterate through the zip

```python
lst1=['krish','sam','john']
lst2=['a','b','c']
```

```python
for i in lst1:
    print(i)
```

```python
output=zip(lst1,lst2)
```

```python
output    
```

```python
type(output)
```

```python
# using iterator for zip object
next(output)
```

```python
list(output)  #typecasting the output elements to the list.
```

```python
# we are getting lst1 and lst2 elements in the form of tupel.
```

```python
output=zip(lst1,lst2)
for i,j in output:
    print(i,j)
```

```python
lst1=['krish','sam','john']
lst2=['a','b','c']
lst3=[1,2,3]
output=zip(lst1,lst2,lst3)
for i,j,k in output:
    print(i,j,k)
```

```python
# creating empty zip
list(zip())
```

```python
# creating dictonary
dict1 = {"name": "Krishna", "age": 20}
dict2 = {"city": "Aurangabad", "country": "India"}
```

```python
dict1        # giving me back the dictionary
```

```python
dict1.items()         # giving back me list
```

```python
dictionary=zip(dict1.items(),dict2.items())
```

```python
for (i,j),(i1,j1) in dictionary:
    print(i,j)
    print(i1,j1)
```

```python
# creating the dictionary form the list using zip
keys1=['name','age']
values1=['krishna',19]
# maing dict
dict1=dict(zip(keys1,values1))

keys2=['city','sex']
values2=['aura','male']
dict2=dict(zip(keys2,values2))
```

```python
print(dict1)
print(dict2)
```
