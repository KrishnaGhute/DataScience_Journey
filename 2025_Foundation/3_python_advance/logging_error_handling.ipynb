```python
import logging

logging.basicConfig(
    filename="krish.txt",
    filemode='a',
    format='%(asctime)s %(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%ML%S'
)

for i in range(0,15):
    if(i%2==0):
        logging.warning("log warning message")
    elif(i%3==0):
        logging.critical("log Critical Message")
    else:
        logging.error("log error message")
```

```python
import logging

# Remove old handlers (important in Jupyter)
for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    level=logging.DEBUG,                      # ðŸ‘ˆ Required to see all logs
    filename="krish.txt",
    filemode='a',
    format='%(asctime)s %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

# Also show logs in Jupyter output
console = logging.StreamHandler()
console.setLevel(logging.DEBUG)
formatter = logging.Formatter('%(asctime)s %(levelname)s - %(message)s')
console.setFormatter(formatter)
logging.getLogger('').addHandler(console)

for i in range(0, 15):
    if i % 2 == 0:
        logging.warning("log warning message")
    elif i % 3 == 0:
        logging.critical("log Critical Message")
    else:
        logging.error("log error message")
```
