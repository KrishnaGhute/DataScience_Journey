```python
# python program for random selection of students from class SYDS
import numpy as np
from numpy import random
import pandas as pd
# created clean and sorted list of boys name, roll number and gender
print("name and roll number of BOYS SYDS:")
bl=pd.read_csv("Boys_List.csv")
print(bl)
print('\n')

# created clean and sorted list of girls name, roll number and gender
print("name and roll number of GIRLS SYDS:")
gl=pd.read_csv("Girls_List.csv")
print(gl)
print('\n')

# boys roll number taken from CSV file and converted to list
roll_num_boys=bl['Roll No'].tolist()
print("roll number of all the boys:",roll_num_boys)
print('\n')

# selecting random boys roll number
x=random.choice(roll_num_boys,size=5,replace=False)
print("random roll numbers of boys",x)
# searching randomly selected boys in list
selected_boys=bl[bl['Roll No'].isin(x)]
print('\n')

# girls roll number taken from CSV file and converted to list
roll_num_girls=gl['Roll No'].tolist()
print("roll number of all the girls:",roll_num_girls)
print('\n')

# selecting random girls roll number
y=random.choice(roll_num_girls,size=5,replace=False)
print("random roll numbers of girls",y)
# searching randomly selected girls in list
selected_girls=gl[gl['Roll No'].isin(y)]
print('\n')

# giving final output from the file with name and roll number.
print("final output")
print("selected boys are:")
print(selected_boys)
print('\n')

print("selected girls are:")
print(selected_girls)
```
