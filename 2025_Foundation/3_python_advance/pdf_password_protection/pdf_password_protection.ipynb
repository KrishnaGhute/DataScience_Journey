```python
!pip install PyPDF2

from PyPDF2 import PdfReader, PdfWriter


def protect_pdf(filename):
    # open the current pdf
    file_pdf = PdfReader(filename)
    # object for pdf write
    out_pdf = PdfWriter()
    # copy pages
    for page in file_pdf.pages:
        out_pdf.add_page(page)
        # üîê Apply encryption (VERY IMPORTANT)
        x=eval(input("Enter your password:"))
        out_pdf.encrypt(user_password=x, owner_password=None)
        # save the encrypted PDF
        name=eval(input("Enter new name for pdf:"))
        form=eval(input("Enter type of pdf you want to open:"))
        with open(name, form) as f:
            out_pdf.write(f)
```
