```python
# class having multiple methods(functions)
class playlist:
    def __init__(self,name):
        self.name=name
        self.songs=[]

    def add_song(self,song):
        self.songs.append(song)
        print(f"added:{song}")
    
    def remove_song(self,song):
        if song in self.songs:
            self.songs.remove(song)
            print(f"removed:{song}")

    def show_song(self,count=0):
        print(f"playlist'{self.name}'")
        count+=1
        for song in self.songs:
            print(f"{count}.{song}")
            count+=1

mp=playlist("favorites")
mp.add_song("hassen")
mp.add_song("nasty")
mp.add_song("gata")
mp.show_song()
mp.remove_song("nasty")
mp.show_song()
```

## Multiple constructor problem in python
- if we create a class with multiple constructor then the constructor get overriden by the constructor with extra parameter.

```python
# creating the class with two constructors
class car():
    def __init__(self,windows,doors):      # constructor with two parameters
        self.windows=windows
        self.doors=doors

    def __init__(self,windows,doors,enginetype):        # constructor with three parameters
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
# trying to access the constructor with two parameters
bmw=car(2,4)
```

- getting an error because of the construcotr with 2 parameters get overridden by the constructor with 3 parameters.

## use postional argument to solve this problem.

```python
class car:
    def __init__(self,*args):
        if len(args)==1:
            self.windows=args[0]
        elif len(args)==2:
            self.windows=args[0]
            self.doors=args[1]
        elif len(args)==3:
            self.windows=args[0]
            self.doors=args[1]
            self.enginetype=args[2]
```

```python
bmw=car(2)
bmw.windows
```

```python
audi=car(2,4,"ev")
print(audi.windows,audi.doors,audi.enginetype)
```
