```python
import matplotlib.pyplot as plt
#%matplotlib inline
# using this line we dont havve to write plot.show().
import numpy as np
```

```python
# taking two variables to create a graph
x=np.arange(0,10)
y=np.arange(11,21)
print(x)
print(y)
```

```python
# ploting graph with matplotlib
# using scatter
plt.scatter(x,y,c='r')
# giving name to the axis
plt.xlabel('x axis')
plt.ylabel('y axis')
# title for whole graph
plt.title('graph  in 2D')
#if we want to save figure
plt.savefig('exgrplt.png')
plt.show()
```

```python
# taking square
y=x*x
plt.plot(x,y,'r*',linestyle='dashed',linewidth='2',markersize='12')
plt.title('squares to number graph')
plt.xlabel('number')
plt.ylabel('squares')
plt.show()
```

```python
# uisng subplot function to ptoting graphs
plt.subplot(2,2,1)
plt.plot(x,y,'r+')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
plt.show()
```

```python
# plotting complex functions
x=np.arange(1,10)
y=3*x+5
plt.plot(x,y,'y')
plt.show()
```

```python
# plotting sin graph 
x=np.arange(0,4*np.pi,0.1)
# 0.1 is step size
y=np.sin(x)
plt.title('sine graph')

plt.plot(x,y)
plt.show()
```

```python
# plotting both sin and cosin graphs
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# ploting both graphs in one
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r')
plt.subplot(2,1,1)
plt.plot(x,y_cos,'g--')
plt.title('sin-cos graph')

# making a subplot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cos graph')

#showing the figure
plt.show()
```

```python
# plotting bar graphs
x=[1,3,5,7,9]
y=[11,33,55,77,99]
x1=[2,3,6,8,10]
y1=[2,22,66,88,10]

# using bar function
plt.bar(x,y)
plt.bar(x1,y1,color='g')
plt.title('bar graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

# showing the graph
plt.show()
```

```python
a=np.array([1,2,6,12,16,14,22,28,26,30,37,44,56,58,99])
plt.hist(a,bins=10)
# bins=10 is by default set in compilre 
plt.title('histogram')
plt.show()
```

```python
# box plot
data=[np.random.normal(0,std,100) for std in range(1,4)]
print(type(data))
print(data)
```

```python
plt.boxplot(data,vert=True,patch_artist=True);
plt.show()
```

```python
# pie chart
# Labels (categories of expenses)
labels = ['Rent', 'Groceries', 'Utilities', 'Transport', 'Entertainment', 'Savings']

# Sizes (percentage or amount spent)
sizes = [25000, 8000, 3000, 2000, 1500, 7000]  # in rupees

# Colors (hex or named colors)
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#C2C2F0', '#FFD700']

#exploding the 1st part
explode=(0,0,0,0,0.1,0,)

#plotting the graph
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
#autopct means show the numbers in floting format
plt.axis('equal')
plt.show()
```
