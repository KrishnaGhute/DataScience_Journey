# series
- a series only have a column

```python
a=[1,2,3,4,5]
df=pd.Series(a)
df
```

```python
df=pd.Series(a,index=['x','y','z','w','t'])
df
```

```python
print(df['y'])
```

# data framle 
- data frame means a matrix form.

```python
#key value pairs in terms of dictionary
weight={"day1":54.97,"day2":55.23,"day3":55.56}
df=pd.Series(weight)
df
```

```python
#if we use list in key value pair
calories={
    "calories":[559,605,590],
    "duration":[50,57,52]
}
df=pd.DataFrame(calories)
df
```

```python
print(df.loc[0])
```

```python
print(df.loc[[0,1]])
```

```python
df=pd.DataFrame(calories,index=["d1","d2","d3"])
df
```

```python
print(df.loc["d2"])
```
