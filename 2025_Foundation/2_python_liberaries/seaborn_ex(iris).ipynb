```python
#working with iris data present in seaborn libesry
import seaborn as sns
iris=sns.load_dataset('iris')
print(iris)
```

```python
iris.head()
```

```python
# we will find the data types for all the columns
iris.dtypes
```

```python
#calculating correlation with heat map
correlation_matrix=iris.corr(numeric_only=True)
print(correlation_matrix)
```

```python
# creating a heat map
sns.heatmap(correlation_matrix)
```

# comparable columns respectively x and y
- sepal_length and petal_length 
- sepal_length and petal_width
- petal_length and petal_width

```python
sns.pairplot(data=iris)
```

```python
iris['species'].value_counts()
```

```python
sns.pairplot(iris,hue='species')
```

```python
# creating jointplot for sepal_length and petal_length
sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='scatter')
```

```python
# creating a jointplot for sepal_length and petal_width
sns.jointplot(x='sepal_length',y='petal_width',data=iris,kind='scatter')
```

```python
sns.histplot(data=iris, x='sepal_length', hue='species', kde=True, bins=10)
```

```python
sns.histplot(data=iris,x='sepal_width',hue="species",kde=True,bins=10)
```

```python
sns.countplot(data=iris,x='sepal_length',hue='species',native_scale=True)
```

```python
# barplot for petal_length and petal_width
sns.barplot(x='petal_length', y='petal_width', data=iris,hue='species',native_scale=True)
```

```python
sns.boxplot(x='petal_length',y='petal_width',hue='species',data=iris,native_scale=True,width=3)
```

```python
sns.violinplot(data=iris,x='petal_length',y='petal_width',hue='species',native_scale=True,width=5)
```
