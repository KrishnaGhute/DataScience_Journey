# corellation with heatmap
-a corellation heatmap uses colored cells, typically on a monochromatic scale, to show a 2d correlation matrix (table) between two discrit dimentions or event types. it is very imp feature in selection

```python
# calculate correlation for only numeric columns
corelation_matrix=df.corr(numeric_only=True)
print(corelation_matrix)
```

```python
sns.heatmap(corelation_matrix)
```

# joint plot
- a join plot allows to study the relationship between 2 numeric variables. the centeral charts display their correlatoin. it is usually a scatterplaot, a hexbin plot , 2d histogram or a 2D density plot.
# univariate analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# pairplot
- A "pairs plot" is also known as a scatterplot, in which one variable in the same data row is matched with  another variables value, like this pair plots are just elaboratas on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
# checking the value of females in the restarunt.
df['sex'].value_counts()
```

```python
# if you want to take different parameters other than the integer points.
# you have to use the hue in the parameters.
sns.pairplot(df,hue='sex')
```

```python
#to check the value_count of smoker
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue="smoker")
```

# Displot
- Dist plot helps us to check the distribution of the columns feature.

```python
sns.displot(df['tip'],kde=True)
# on y axis the density is in percentsge
```

```python
# without kde(kernal density estimation)
sns.displot(df['tip'],kde=False,bins=10)
```

# categorical plots
- Seaborn also helps us in doing the analysis on categorical data points. in this section we will discuss about
- boxplot
- violinplot
- countplot
- bar plot

```python
# count plot:Show the counts of observations in each categorical bin using bars.
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
#bar plot:Show point estimates and errors as rectangular bars.
sns.barplot(x='total_bill',y='sex',data=df)
```

# Box plot
- A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,hue='smoker',palette='rainbow')
```

```python
# to take all the parameters
sns.boxplot(data=df,orient="v")
```

```python
# catagorize my data based on some other catogeries
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# voilin plot
voilin plot helps us to see both the distribution of data in terms of kernal density estimation and box plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
